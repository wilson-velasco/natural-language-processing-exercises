1) Codeup Blog Articles

Visit <a href="https://codeup.com/blog/">Codeup's Blog</a> and record the urls for at least 5 distinct blog posts. For each post, you should scrape at least the post's title and content.

Encapsulate your work in a function named `get_blog_articles` that will return a list of dictionaries, with each dictionary representing one article. The shape of each dictionary should look like this:

<code>{
    'title': 'the title of the article',
    'content': 'the full text content of the article'
}</code>

Plus any additional properties you think might be helpful.

**Bonus:** Scrape the text of all the articles linked on codeup's blog page.

````python
#Imports for scraping

import requests
from bs4 import BeautifulSoup
import os
````

````python
#To allow access since otherwise it's a 403 error

headers = {"User-Agent": "Chrome/91.0.4472.124"}
````

````python
#Grabbing some random URLs from the blog

urls = ['https://codeup.com/featured/apida-heritage-month/'
       ,'https://codeup.com/featured/women-in-tech-panelist-spotlight/'
       ,'https://codeup.com/events/black-excellence-in-tech-panelist-spotlight-stephanie-jones/'
       ,'https://codeup.com/codeup-news/codeup-best-bootcamps/'
       ,'https://codeup.com/employers/hiring-tech-talent/']
````

````python
#Double-checking access

for url in urls:
    print(requests.get(url, headers = headers))
````

````python
#Looking at what one response brings for first URL

response = requests.get(urls[0], headers = headers)
response.text
````

````python
#Putting it into HTML parser

soup = BeautifulSoup(response.content, 'html.parser')
````

````python
#Getting title from page

title = soup.title.string

title
````

````python
#For-looping to get all text from article (this wasn't as effective as just accessing div entry-content. See below.)

for element in soup.find_all('p'):
    print(element.text)
````

````python
urls
````

````python
#Putting everything above in one coded function

code_up_dict = []

for url in urls:
    response = requests.get(url, headers = headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    #Get title
    title = soup.title.string
    
    #Get content
    contents = soup.find('div', class_='entry-content').text.replace('\n', ' ')
    
    dict_entry = {'url': [url]
                ,'title': [title]
                ,'content': [contents]}
        
    code_up_dict.append(dict_entry)

code_up_dict
````

2) News Articles

We will now be scraping text data from <a href="https://inshorts.com">inshorts</a>, a website that provides a brief overview of many different topics.

Write a function that scrapes the news articles for the following topics:

- Business
- Sports
- Technology
- Entertainment

The end product of this should be a function named `get_news_articles` that returns a list of dictionaries, where each dictionary has this shape:

```{
    'title': 'The article title',
    'content': 'The article content',
    'category': 'business' # for example
}```

Hints:

a) Start by inspecting the website in your browser. Figure out which elements will be useful.

b) Start by creating a function that handles a single article and produces a dictionary like the one above.

c) Next create a function that will find all the articles on a single page and call the function you created in the last step for every article on the page.

d) Now create a function that will use the previous two functions to scrape the articles from all the pages that you need, and do any additional processing that needs to be done.

````python
#Trying it out on just one URL so far

inshorts_url = 'https://inshorts.com/en/read/business'
````

````python
#Make the soup

response = requests.get(inshorts_url)

soup = BeautifulSoup(response.content, 'html.parser')
````

````python
#Checking to see HTML content per card, looking at the first card

soup.find('div', class_='news-card z-depth-1')
````

````python
#Getting title from card

soup.find('span', itemprop='headline').text
````

````python
#Getting text from the article from the card

(soup.find('div', itemprop='articleBody').text).replace('\\','')
````

````python
#Putting all of the above in one function

categories = ['business', 'sports', 'technology', 'entertainment']

inshorts_dict = []

for categorie in categories:
    url = f'https://inshorts.com/en/read/{categorie}'
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    for element in soup.find_all('div', class_='news-card z-depth-1'):
        title = element.find('span', itemprop='headline').text
        content = element.find('div', itemprop='articleBody').text.replace('\n', '')
        category = categorie
        
        dict_entry = {'title': title
                     ,'content': content
                     ,'category': category}
        
        inshorts_dict.append(dict_entry)

inshorts_dict
````
